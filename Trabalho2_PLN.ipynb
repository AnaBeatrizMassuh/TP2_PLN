Fução para remover padrões

```python
def remove_pattern(input_text, pattern):
  r = re.findall(pattern, input_text)
  for word in r:
    input_text = re.sub(word, "", input_text)

  return input_text 
```

Remover os rt

```python
df['clean_text'] = np.vectorize(remove_pattern)(df['clean_text'], 'rt[\s]+') 
df
```

Remover @usuarios

```python
df['clean_text'] = np.vectorize(remove_pattern)(df['clean_text'], '@[A-Za-z0-9]+') 
df
```

Remover links sites

```python
df['clean_text'] = np.vectorize(remove_pattern)(df['clean_text'], 'http?:\/\/') 
df
```

Remover caracteres especiais, pontuação , numeros e hashtag 

```python
df['clean_text'] = df['clean_text'].str.replace('[^A-Za-z]', ' ') 
df
```

Retirar stopwords

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
stop_words
```

```python
df['filtered_text'] = df['clean_text'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
df
```

Retirar lemmatização

```python
nltk.download('omw-1.4')
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
df['filtered_text'] = df['filtered_text'].apply( lambda x: ' '.join([lemmatizer.lemmatize(word) for word in x.split()]))
df
```

Retirar a stemmização

```python
from nltk.stem.porter import *
stemmer = PorterStemmer()
df['filtered_text'] = [stemmer.stem(t) for t in df['filtered_text']]
df
```
